```python
import json
import pandas as pd

baseurl = "https://knightconnect.campuslabs.com/engage/organizations/"
pictureurl = "https://se-images.campuslabs.com/clink/images/"
```

```python
# Cleaning the data from rsos.json writing to rsos.txt
with open('rsos.json', 'r') as f:
    rso_data = json.load(f)

df = pd.DataFrame(columns=['name', 'website', 'image', 'description', 'summary', 'categoryNames'])

for item in rso_data['value']:
    df.loc[len(df)] = [item['Name'], baseurl + item['WebsiteKey'], pictureurl + item['ProfilePicture'] if item['ProfilePicture'] else None,item['Description'], item['Summary'], item['CategoryNames']]

df.to_csv('rsos.csv', index=False)
```

```python
# Cleaning the data from rso_events.json writing to rsos.txt
with open('../rso_events.json', 'r') as f:
    event_data = json.load(f)

events = []
# organizationName, organizationNames, description, location, startsOn endsOn, categoryNames
df = pd.DataFrame(columns=['organizationName', 'organizationNames', 'description', 'location', 'startsOn', 'endsOn', 'categoryNames'])
for item in event_data['value']:
    df.loc[len(df)] = [item.get('organizationName', ''), item.get('organizationNames', ''),
                       item.get('description', ''), item.get('location', ''),
                       item.get('startsOn', ''), item.get('endsOn', ''),
                       item.get('categoryNames', '')]
df['startsOn'] = pd.to_datetime(df['startsOn'], errors='coerce')
df['endsOn'] = pd.to_datetime(df['endsOn'], errors='coerce')
df.to_csv('../rso_events.csv', index=False)
```
